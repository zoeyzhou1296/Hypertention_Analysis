## filter by the missing values of each
80% of varaibles for each rows should be available

```python
Hyper_master['na_count'] = Hyper_master.isnull().sum(axis=1)
```

```python
Hyper_master_final = Hyper_master[Hyper_master['na_count'] < 57]
```

```python
col_list=list(Hyper_master_final)
out_list=[]
for col_name in col_list:
  num = Hyper_master_final[col_name].isnull().sum(axis=0)
  if num > 1000:
    out_list.append(col_name)
out_list
```

```python
Hyper_master_final=Hyper_master_final.drop(columns=out_list)
```

```python
Hypertension = pd.read_csv(BASE_PATH + '/Hypertension_final.csv', index_col=0)
Hyper_master_final = Hyper_master_final.merge(Hypertension, on = 'SEQN', how = 'left')
```
