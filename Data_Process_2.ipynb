```python
def merge(year):
  BASE_PATH = "/content/drive/MyDrive/Hypertension_Final_Project/cleaned_file/" + str(year)
  
  entries = [entry for entry in os.listdir(BASE_PATH) if os.path.isfile(os.path.join(BASE_PATH, entry))]
  merge_list=entries[2:]
  Hyper_master = pd.read_csv(BASE_PATH + '/Hyper_SEQN.csv', index_col=0)
  Hypertension = pd.read_csv(BASE_PATH + '/Hypertension_final.csv', index_col=0)
  Hyper_master = Hyper_master.merge(Hypertension, on = 'SEQN', how = 'left')
  for file in merge_list:
    file_read = pd.read_csv(BASE_PATH + '/' + file, index_col=0)
    Hyper_master = Hyper_master.merge(file_read, on = 'SEQN', how = 'left')
  return Hyper_master
```

```python
for i in range(2007, 2019, 2):
    hyper_master_temp = merge(i)
    if i == 2007:
      hyper_master = hyper_master_temp
    else:
      hyper_master = pd.concat([hyper_master, hyper_master_temp], ignore_index=True)
```
